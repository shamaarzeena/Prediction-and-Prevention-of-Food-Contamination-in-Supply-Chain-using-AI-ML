```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils.class_weight import compute_class_weight
```

```python
def create_mlp(input_dim):
    model = Sequential()
    model.add(Dense(64, input_dim=input_dim, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(32, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(16, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))  # Binary classification output
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
input_dim = X_train_scaled.shape[1]
model = create_mlp(input_dim)

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    X_train_scaled, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stop],
    verbose=2

)
```

```python
# Predict probabilities and classes
y_pred_proba = model.predict(X_test_scaled).flatten()
y_pred = (y_pred_proba > 0.5).astype(int)

# Metrics
print("MLP Accuracy:", accuracy_score(y_test, y_pred))
print("MLP Precision:", precision_score(y_test, y_pred))
print("MLP Recall:", recall_score(y_test, y_pred))
print("MLP F1 Score:", f1_score(y_test, y_pred))
print("MLP ROC AUC:", roc_auc_score(y_test, y_pred_proba))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("MLP Confusion Matrix")
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, LSTM, Dense, Dropout, Input, Bidirectional
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import backend as K
from tensorflow.keras.layers import GlobalAveragePooling1D


# Focal Loss
def focal_loss(gamma=2., alpha=0.25):
    def focal_loss_fixed(y_true, y_pred):
        epsilon = K.epsilon()
        y_pred = K.clip(y_pred, epsilon, 1. - epsilon)
        cross_entropy = -y_true * K.log(y_pred) - (1 - y_true) * K.log(1 - y_pred)
        weight = alpha * K.pow(1 - y_pred, gamma) * y_true + (1 - alpha) * K.pow(y_pred, gamma) * (1 - y_true)
        loss = weight * cross_entropy
        return K.mean(loss)
    return focal_loss_fixed

# Model
from tensorflow.keras import regularizers

model = Sequential([
    Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.3),
    Conv1D(32, kernel_size=3, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    GlobalAveragePooling1D(),
    Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


model.summary()
```

```python
# ===========================
# STEP 3: Clustering to uncover risk phenotypes
# ===========================
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# If X_test_scaled is NumPy, convert to DataFrame
n_features = X_test_scaled.shape[1]
feature_names = [f"f{i}" for i in range(n_features)]
X_test_df = pd.DataFrame(X_test_scaled, columns=feature_names)

# Standardize continuous features for clustering
scaler = StandardScaler()
X_test_scaled_std = scaler.fit_transform(X_test_df)

# Try k = 2..6, pick best Silhouette
best_k, best_score = None, -1
for k in range(2, 7):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels_k = km.fit_predict(X_test_scaled_std)
    score = silhouette_score(X_test_scaled_std, labels_k)
    if score > best_score:
        best_score, best_k = score, k

# Final clustering
k = best_k if best_k is not None else 3
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_test_scaled_std)

# Build results DataFrame
results = pd.DataFrame({
    "y_true": y_test,
    "y_prob": xgb_proba,
    "y_pred": xgb_preds,
    "cluster": cluster_labels
})

print(f"Chosen k: {k}, Silhouette: {round(best_score,3)}")

# Cluster summary
cluster_summary = results.groupby("cluster").agg(
    n=("y_true","count"),
    avg_prob=("y_prob","mean"),
    pos_rate=("y_true","mean"),
    acc=("y_true", lambda a: (a.values == results.loc[a.index, "y_pred"].values).mean())
).reset_index()

print(cluster_summary.sort_values("avg_prob", ascending=False))

# Map clusters to risk tiers
def tier_from_prob(p):
    if p >= 0.6: return "High"
    if p >= 0.3: return "Medium"
    return "Low"

cluster_summary["risk_tier"] = cluster_summary["avg_prob"].apply(tier_from_prob)
results = results.merge(cluster_summary[["cluster","risk_tier"]], on="cluster", how="left")

print("Step 3 completed: Clusters labeled with risk tiers.")
```

```python
# --- Build X_clust from X_shap and run KMeans + heatmap ---

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# 0) Ensure X_shap exists and has your real feature names as columns
assert isinstance(X_shap, pd.DataFrame), "X_shap must be a DataFrame with named columns"

# 1) Choose features for clustering (mostly continuous + integrity/flags)
preferred_cols = [
    'Temperature_C','Humidity_Percent','pH_Level','Gas_Concentration_ppm',
    'Vibration_Level','Light_Exposure_Lux','Handling_Score','Time_in_Stage_hr',
    'Ambient_Air_Quality','Container_Seal_Integrity','Cleanliness_Index',
    'Power_Stability_Flag','Previous_Contamination_Flag','Supply_Delay_Flag'
]

# keep only columns that actually exist (intersection)
cols = [c for c in preferred_cols if c in X_shap.columns]
assert len(cols) >= 6, f"Too few columns for clustering. Found: {cols}"

# 2) Standardize (fit on the data you’re clustering; using test set is fine for the figure)
scaler = StandardScaler()
X_clust = scaler.fit_transform(X_shap[cols])

# 3) Pick k via Silhouette over k=2..6
k_candidates = list(range(2, 7))
sil_scores = []
for k in k_candidates:
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = km.fit_predict(X_clust)
    sil_scores.append(silhouette_score(X_clust, labels))

best_k = k_candidates[int(np.argmax(sil_scores))]
kmeans = KMeans(n_clusters=best_k, n_init=10, random_state=42)
labels = kmeans.fit_predict(X_clust)

print(f"Chosen k = {best_k}; silhouette scores by k:", dict(zip(k_candidates, sil_scores)))

# 4) Compute cluster "means" in standardized space (comparable across features)
X_std = pd.DataFrame(X_clust, columns=cols)
X_std["Cluster"] = labels
cluster_means = X_std.groupby("Cluster").mean().sort_index()

# 5) Plot and save heatmap with real feature names
outdir = Path("figs_shap_relabelled")
outdir.mkdir(exist_ok=True)

plt.figure(figsize=(1.2*len(cols), 0.8*best_k + 4))
ax = sns.heatmap(
    cluster_means[cols],
    cmap="coolwarm",
    center=0.0,
    xticklabels=cols,
    yticklabels=[f"Cluster {i}" for i in cluster_means.index]
)
ax.set_xlabel("Features")
ax.set_ylabel("Cluster")
ax.set_title(f"Cluster Feature Means (standardized), k={best_k}")
plt.tight_layout()
plt.savefig(outdir/"cluster_feature_heatmap.png", dpi=220)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 3.1 risk tiers from probabilities (you can tune thresholds)
def risk_tier(p):
    if p >= max(0.8, chosen_threshold + 0.15):  # high
        return "High"
    elif p >= chosen_threshold:                 # medium
        return "Medium"
    else:
        return "Low"

risk_tiers = np.array([risk_tier(p) for p in xgb_proba_aligned])

# 3.2 prepare continuous matrix for clustering (use standardized engineered features)
X_for_cluster = np.asarray(Xvalues, dtype=float)
Xc = StandardScaler().fit_transform(X_for_cluster)

# 3.3 choose k via silhouette on k=2..6
best_k, best_s, best_model = None, -1, None
for k in range(2, 7):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(Xc)
    s = silhouette_score(Xc, labels)
    if s > best_s:
        best_s, best_k, best_model = s, k, km

cluster_labels = best_model.predict(Xc)

# 3.4 cluster profiling
df_export = pd.DataFrame({
    "batch_id": Xts_aligned.index if hasattr(Xts_aligned, "index") else np.arange(Xvalues.shape[0]),
    "y_true": np.array(yts_aligned.values),
    "y_proba_xgb": xgb_proba_aligned,
    "y_pred_xgb": (xgb_proba_aligned >= chosen_threshold).astype(int),
    "risk_tier": risk_tiers,
    "cluster": cluster_labels
})

profile_rows = []
for c in sorted(np.unique(cluster_labels)):
    mask = cluster_labels == c
    profile_rows.append({
        "cluster": int(c),
        "n": int(mask.sum()),
        "mean_proba": float(df_export.loc[mask, "y_proba_xgb"].mean()),
        "pos_rate": float(df_export.loc[mask, "y_true"].mean()),
        "risk_high_pct": float((df_export.loc[mask, "risk_tier"]=="High").mean())
    })
cluster_profile = pd.DataFrame(profile_rows).sort_values("mean_proba", ascending=False)

df_export.to_csv("predictions_with_tiers_clusters.csv", index=False)
cluster_profile.to_csv("cluster_profile.csv", index=False)

print(f"Chosen k: {best_k} (silhouette={best_s:.3f}) — saved predictions_with_tiers_clusters.csv & cluster_profile.csv")
```
